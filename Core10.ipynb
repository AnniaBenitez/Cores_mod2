```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
test = pd.read_csv("test.csv")
test.head()
```

```python
# Calcular el total de valores faltantes en cada columna y ordenarlos de forma descendente
total = train.isnull().sum().sort_values(ascending=False)

# Calcular el porcentaje de valores faltantes en cada columna y ordenarlos de forma descendente
percent = (train.isnull().sum() / train.isnull().count()).sort_values(ascending=False)

# Combinar ambas métricas en un solo DataFrame
missing_data = pd.concat([total, percent], axis=1, keys=['Cantidad de faltantes', 'Porcentaje de faltantes'])

# Mostrar el resultado
missing_data
```

```python
train.Name = train.Name.fillna('Desconocido')
print(train.isnull().sum())

test.Name = test.Name.fillna('Desconocido')
print(test.isnull().sum())
```

```python
train['CryoSleep']=train['CryoSleep'].fillna(False)
train['VIP']=train['VIP'].fillna(False)
print(train.isnull().sum())

test['CryoSleep']=test['CryoSleep'].fillna(False)
test['VIP']=test['VIP'].fillna(False)
print(test.isnull().sum())
```

```python
train['Age']=train['Age'].fillna(train['Age'].mean())
print(train.isnull().sum())

test['Age']=test['Age'].fillna(test['Age'].mean())
print(test.isnull().sum())
```

```python
train['RoomService']=train['RoomService'].fillna(train['RoomService'].mean())
train['FoodCourt']=train['FoodCourt'].fillna(train['FoodCourt'].mean())
train['ShoppingMall']=train['ShoppingMall'].fillna(train['ShoppingMall'].mean())
train['Spa']=train['Spa'].fillna(train['Spa'].mean())
train['VRDeck']=train['VRDeck'].fillna(train['VRDeck'].mean())
print(train.isnull().sum())

test['RoomService']=test['RoomService'].fillna(test['RoomService'].mean())
test['FoodCourt']=test['FoodCourt'].fillna(test['FoodCourt'].mean())
test['ShoppingMall']=test['ShoppingMall'].fillna(test['ShoppingMall'].mean())
test['Spa']=test['Spa'].fillna(test['Spa'].mean())
test['VRDeck']=test['VRDeck'].fillna(test['VRDeck'].mean())
print(test.isnull().sum())
```

```python
train['Destination']=train['Destination'].fillna('TRAPPIST-1e')
train['HomePlanet']=train['HomePlanet'].fillna('Earth')
print(train.isnull().sum())
test['Destination']=test['Destination'].fillna('TRAPPIST-1e')
test['HomePlanet']=test['HomePlanet'].fillna('Earth')
print(test.isnull().sum())
```

```python
train['Cabin'] =train['Cabin'].fillna('T/0/P')
test['Cabin'] =test['Cabin'].fillna('T/0/P')
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
# Crear una lista desde cero con los valores deseados
high_corr_list = ['CryoSleep', 'FoodCourt', 'ShoppingMall', 'VIP', 'Age']

# Mostrar la lista
print(high_corr_list)
```

```python
from sklearn.model_selection import train_test_split
X = train[high_corr_list]
y = train['Transported']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
knn_pred = knn.predict(X_test)
print("KNeighbors Classifier accuracy: {}".format(metrics.accuracy_score(y_test,knn_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=1000)
rfc.fit(X_train,y_train)
rfc_pred = rfc.predict(X_test)
print("Random forest accuracy: {}".format(metrics.accuracy_score(y_test,rfc_pred)))
```

Random forest tiene mejor accuracy
